```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=["Donate"])
Y = df["Donate"]
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.6, 
    random_state=42,stratify=df["Donate"])
```

```python
print(X_train.shape,X_test.shape)
```

```python
import joblib
#from sklearn.tree import DecisionTreeClassifier
#from sklearn.neighbors import KNeighborsClassifier
#from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import RandomForestClassifier
import os
```

```python
model=RandomForestClassifier(n_estimators=38, random_state=42)#(n_estimators= 38) 
#model = KNeighborsClassifier(n_neighbors=10)
#model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
file_path = 'decision_tree_model.pkl'

try:
    # Save the trained model to the file path
    joblib.dump(model, file_path)
    # Check if the file exists
    if os.path.exists(file_path):
        print("File saved successfully to the same folder as the Python file.")
    else:
        print("Error: File was not saved.")
except Exception as e:
    print(f"Error: {e}")
```
